### CONCATENATION, JOINING

```python
df1 = pd.DataFrame( ((1, 2, 3), (4, 5, 6)), columns = list("abc"))
df2 = pd.DataFrame( ((10, 11, 12),(13, 14, 15)), columns = list("abc"))
```

```python
print(df1)
print(df2)
```

```python
pd.concat((df1, df2), ignore_index = True)
```

```python
pd.concat((df1, df2))
```

```python
df1 = pd.DataFrame(((1, 2, 3), (4, 5, 6)), columns = list("abc"), index = (1, 2))
df2 = pd.DataFrame(((10, 11, 12), (13, 14, 15)), columns = list("xyz"), index = (10, 2))                  
```

```python
df1
```

```python
df2
```

```python
pd.concat((df1, df2), axis = 1)  # JOIN BASED ON ROW INDEX - OUTER JOIN
```

```python
pd.concat((df1, df2))   
```

```python
pd.concat((df1, df2), axis = 1, join = "inner")
```

```python
df1 = pd.DataFrame({ "name" : ["A", "B", "C"], "age" : [25, 20, 26]})
df2 = pd.DataFrame({ "name" : ["A", "B", "D"], "gender" : ["M", "F", "F"]})
```

```python
df1
```

```python
df2
```

```python
pd.concat((df1, df2), axis = 1)
```

```python
df1.merge(df2) # MERGE BASED ON COMMON COLUMN - NAME (INNER JOIN)
```

```python
df1.merge(df2, on = "name", how = "outer")
```

```python
df1.merge(df2, left_index = True, right_index = True)  # MERGE BASED ON ROW INDEXES
```

```python
df1 = pd.DataFrame( {"name" : ["A", "B", "C"], "age" : [25, 20, 26]})
df2 = pd.DataFrame( {"firstname" : ["A", "B", "D"], "gender" : ["M", "F", "F"]})
```

```python
df1
```

```python
df2
```

```python
# WHEN COLUMN NAMES ARE NOT SAME USE LEFT_ON AND RIGHT_ON
df1.merge(df2)
```

```python
df1.merge(df2, left_on = "name", right_on = "firstname", how = "inner")
```

```python
df1.merge(df2, left_on = "name", right_on = "firstname", how = "outer")
```
