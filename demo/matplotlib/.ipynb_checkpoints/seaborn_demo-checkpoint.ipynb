```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set(style = "dark")
```

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

# CATEGORICAL PLOTS

### CATEGORICAL PLOTS

### box plot, cat plot, count plot

```python
ax = sns.boxplot(x = "day", y = "tip", data = tips)
```

```python
ax = sns.boxplot(x = "day", y = "tip", showfliers = False, data = tips)
```

```python
ax = sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = tips)
ax.set_xlabel("Week")
t = ax.set_title("Day-wise Total Bill for Male and Female") 
```

```python
flights.info()
```

```python
flights.describe()
```

```python
ax = sns.boxplot(x = "day", y = "tip", hue = "smoker", data = tips)
```

```python
# ADD EXTRA PARAMETERS RELATED TO MATPLOTLIB
plt.figure(figsize = (10, 5))
ax = sns.boxplot(x = "day", y = "total_bill", hue = "sex", showmeans = True, showfliers = False, width = 0.5, data = tips)
```

```python
plt.figure(figsize = (10, 5))
ax = sns.boxplot(x = "day", y = "total_bill", hue = "sex", showmeans = True, meanline = True,
                 showfliers = False, width = 0.5, data = tips)
```
