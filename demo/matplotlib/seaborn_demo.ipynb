# MATRIX PLOTS

### HEATMAP

```python
day_tips = tips.groupby("day", as_index = False)[["tip", "total_bill"]].sum()
```

```python
day_tips
```

```python
ds = tips.pivot_table(index = "day", columns = "time", values = "total_bill", aggfunc = sum)
```

```python
ds
```

```python
sns.heatmap(ds)    # HEAT MAP
```

```python
sns.heatmap(ds, cmap = "YlGnBu", annot = True, fmt = ".2f")
```

```python
fl_summary = flights.pivot("month", "year", "passengers")
print(fl_summary)
ax = sns.heatmap(fl_summary, annot = True, fmt = ".2f")
plt.gcf().set_size_inches(20, 15)
```

```python
fl_summary = flights.pivot('month','year','passengers')
ax = sns.heatmap(fl_summary,annot=True, fmt=".0f", cbar=False)
plt.gcf().set_size_inches(20,10)
```

```python
cm = tips[["total_bill", "tip", "size"]].corr()
print(cm)
sns.heatmap(cm)
```

```python
cm = tips[["total_bill", "tip", "size"]].corr()
print(cm)
sns.heatmap(cm, annot = True)
```
